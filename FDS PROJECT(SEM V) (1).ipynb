```python
df.drop_duplicates(inplace=True)
df
```

STEP 4 :- Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.hist(df['Age'],color="blue")
plt.show()
```

```python
sns.histplot(df,x="Age")
plt.show()
```

```python
#histogram for a specific feature
plt.figure(figsize=(10,7))
plt.hist(df["BP"])
plt.xlabel("BP")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Comparison between Chest pain type based on BP and Cholesterol')
sns.scatterplot(x=df['BP'],y=df['Cholesterol'],hue=df['Chest pain type'],s=50)
```

```python
plt.title("Max HR")
sns.boxplot(df["Max HR"],color='orange')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.scatter3D(df['BP'],df['Cholesterol'],df['FBS over 120'],edgecolor='black');
ax.set_xlabel("BP")
ax.set_ylabel("Cholesterol")
ax.set_zlabel("FBS over 120")
```
